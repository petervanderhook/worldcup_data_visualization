```python
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd
import datetime


df2 = pd.read_csv('worldcup_clean.csv')

today = datetime.datetime.now()
age_dict = {}

age = datetime.datetime.now() - datetime.datetime.strptime(df2['DOB'][0], '%Y-%m-%d')
today_date = datetime.datetime.strptime(f"{datetime.datetime.now().year}-{datetime.datetime.now().month}-{datetime.datetime.now().day}", '%Y-%m-%d')
for count, dob in enumerate(df2['DOB']):
    group_num = df2['Group'][count]
    player_dob = datetime.datetime.strptime(dob, '%Y-%m-%d')
    age = today_date.year - player_dob.year
    if not group_num in age_dict:
        age_dict[group_num] = []
        age_dict[group_num].append(age)
    else:
        age_dict[group_num].append(age)
mean_list = []
for i in age_dict:
    m = np.mean(age_dict[i])
    #print(len(age_dict[i]))
    #print(f"\n\n{i}: {len(age_dict[i])}\nMean: {m}")
    mean_list.append(m)
plt.hist(mean_list, bins=[32, 32.25, 32.5, 32.75, 33, 33.25, 33.5, 33.75, 34])
plt.xlabel("Mean of Group's Age")
plt.ylabel("Count of Means")
plt.title("Histogram of the mean of the age for each group.")
```
